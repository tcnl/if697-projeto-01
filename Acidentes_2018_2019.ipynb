```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
import seaborn as sns
import numpy as np
import datetime
```

# Pré processamento

```
#sem ajustes para importar o csv
urls_acidentes = ['https://raw.githubusercontent.com/tcnl/tcnl/master/acidentes_2018.csv', 'https://raw.githubusercontent.com/tcnl/tcnl/master/acidentes-2019.csv']

dados_acidentes = pd.concat((pd.read_csv(u, sep=';') for u in urls_acidentes), axis=0, ignore_index=True)
```

```
dados_acidentes = dados_acidentes.filter(['DATA', 'hora', 'natureza_acidente', 'bairro', 'tipo', 'condicao_via'])

dados_acidentes
```

```
dados_acidentes.dtypes
```

```
#extrair dia da semana e ano
data_acidente = pd.to_datetime(dados_acidentes['DATA'], format="%Y-%m-%d", errors='coerce')

dados_acidentes['dia_da_semana'] = data_acidente.apply(lambda x: x.weekday())
dados_acidentes['ano'] = data_acidente.apply(lambda x: x.year)

dados_acidentes
```

```
#extrair hora
dados_acidentes['hora_bruta'] = pd.to_datetime(dados_acidentes['hora'], format="%H:%M:%S", errors='coerce').apply(lambda x: x.hour)

dados_acidentes
```

```
#remove as colunas de data e hora
dados_acidentes.drop(columns=['DATA', 'hora'], inplace=True)
dados_acidentes
```

```
#limpeza dos dados
dados_acidentes['bairro'].replace(['BOMBA DO HEMETERIO', 'JOANA BEZERRA', 'IPESEP'], ['BOMBA DO HEMETÉRIO', 'ILHA JOANA BEZERRA', 'IPSEP'], inplace=True)
dados_acidentes['natureza_acidente'].replace(['APOIO', 'ENTRADA E SAÍDA'], [np.nan, np.nan], inplace=True)
dados_acidentes['tipo'].replace(['SUBURBIO', '0', 'SANTO AMARO'], [np.nan, np.nan, np.nan], inplace=True)
```

```
dados_acidentes
```

```
#tratamento de NA e categorização

categories_list = ['natureza_acidente', 'bairro', 'tipo', 'condicao_via']

dados_acidentes['natureza_acidente'].fillna('SEM VÍTIMA', inplace=True)
dados_acidentes = dados_acidentes.apply(lambda x: x.fillna('Não informado') if x.name in categories_list else x)
dados_acidentes = dados_acidentes.apply(lambda x: x.astype('category') if x.name in categories_list else x)

dados_acidentes
```

```
#categorias em números

dados_acidentes = dados_acidentes.apply(lambda x: x.cat.codes if x.name in categories_list else x)
dados_acidentes
```

```
dados_acidentes.dtypes
```

```
dados_acidentes = dados_acidentes.convert_dtypes()
dados_acidentes.dtypes
```

```
dados_acidentes
```

```
dados_acidentes.to_csv('./acidentes_2018_2019_pre_processado.csv')
```
